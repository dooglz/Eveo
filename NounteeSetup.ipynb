```python
def _makey(noun,qty,order):
    if type(noun) == type(dict()):
        return (order,[noun])
    return (order,[{noun:qty}])

def Buy(noun='icecream',qty=6000.0):
    return _makey(noun,qty,"BUY")
    #return ("BUY",pi.ItemQty(nounIds[noun],qty))

def Make(noun='icecream',qty=6000.0):
    return _makey(noun,qty,"MAKE")
    #return  ("Make",pi.ItemQty(nounIds[noun],qty))

def Sell(noun='icecream',qty=6000.0):
    return _makey(noun,qty,"SELL")
    #return  ("Sell",pi.ItemQty(nounIds[noun],qty))
```
